Albert Gong was selected as one of top 300 STS scholar for 2019.

This notebook will convert the STS-book from pdf into csv, so one can easily browse data

```python
# get a list of US state names from web
# save to text file="usa-states.txt"  (one per line)
# store them in mapState {key:state_name, value:line_number}
fo = open("usa-states.txt")
all_states = [l.strip().upper() for l in fo.readlines()]
fo.close()
print(all_states)
```

```python
# add missing 'DISTRICT OF COLUMBIA'
all_states = ['ALABAMA', 'ALASKA', 'ARIZONA', 'ARKANSAS', 'CALIFORNIA', 'COLORADO', 'CONNECTICUT', 'DELAWARE', 'DISTRICT OF COLUMBIA', 'FLORIDA', 'GEORGIA', 'HAWAII', 'IDAHO', 'ILLINOIS', 'INDIANA', 'IOWA', 'KANSAS', 'KENTUCKY', 'LOUISIANA', 'MAINE', 'MARYLAND', 'MASSACHUSETTS', 'MICHIGAN', 'MINNESOTA', 'MISSISSIPPI', 'MISSOURI', 'MONTANA', 'NEBRASKA', 'NEVADA', 'NEW HAMPSHIRE', 'NEW JERSEY', 'NEW MEXICO', 'NEW YORK', 'NORTH CAROLINA', 'NORTH DAKOTA', 'OHIO', 'OKLAHOMA', 'OREGON', 'PENNSYLVANIA', 'RHODE ISLAND', 'SOUTH CAROLINA', 'SOUTH DAKOTA', 'TENNESSEE', 'TEXAS', 'UTAH', 'VERMONT', 'VIRGINIA', 'WASHINGTON', 'WEST VIRGINIA', 'WISCONSIN', 'WYOMING', 'CANADA', 'CHINA']
```

```python
mapState = {}
for l in all_states:
    state = l.upper().strip()
    #print(state)
    mapState[state] = -1

mapState.keys()
```

```python
# text file "STS-books-2019__2.txt" is copy-and-pasted out from the PDF
# set line_number for each state

fo2 = open("STS-books-2019__2.txt")
alllines = fo2.readlines()
fo2.close()

iline = 0
for l in alllines:
    tmp = l.upper().strip()
    if tmp in mapState.keys():
        mapState[tmp] = iline

    iline += 1
```

```python
# if a state has value = -1, it means no student from that state was selected as scholar
print(mapState)
```

```python
# store State into a list maintaining its order in PDF
lstState = []
for k,v in mapState.items():
    #print(k,v)
    if v == -1:
        continue
    lstState.append(k)
```

```python
print(lstState)
```

```python
# mapState2 tracks end-of-line-number for a given state
mapState2 = {}
for i in range(len(lstState)):
    if i == len(lstState)-1:
        mapState2[lstState[i]] = len(alllines)
    else:
        mapState2[lstState[i]] = mapState[lstState[i+1]]
```

```python
print(mapState2)
```

```python
# split original text file by state
for state in mapState2.keys():
    filename = "%s.txt" % (state,)
    fo4 = open(filename, "w")
    fo4.write("".join(alllines[(mapState[state]+1):mapState2[state]]))
fo4.close()
```

```python
# detect and parse the line with student name/age
def match_student(student):
    import re
    matchObj = re.match(r"(.*), (.*), (\d+)", student)
    if matchObj:
       #print ("matchObj.group() : ", matchObj.group())
       #print ("matchObj.group(1) : ", matchObj.group(1))
       #print ("matchObj.group(2) : ", matchObj.group(2))
       #print ("matchObj.group(3) : ", matchObj.group(3)) 
        return [matchObj.group(1), matchObj.group(2), matchObj.group(3)]
    else:
        return [student]
```

```python
# map to store project_title, key="state#school#student" 
mapProject = {}

for state in mapState2.keys():
    #state = "FLORIDA"
    filename = "%s.txt" % (state,)
    fo5 = open(filename)
    data = fo5.readlines()
    fo5.close()

    #data

    i = -1
    for d in data:
        i += 1
        if len(d.strip()) < 1:
            continue  # skip blank line
        if "SCHOOL" in d.upper() or "ACADEM" in d.upper():
            school = d.strip()
            #mapSchool[school] = i
            continue
        student_info = match_student(d)
        if len(student_info) == 3:
            student = "_".join(student_info)
            #mapStudent[student] = i
            key = state+"#"+school+"#"+student
            mapProject[key] = []
            continue
        project = d.strip()
        print("key= ", key)
        mapProject[key].append(project)
```

```python
# write out parsed records into csv
fo6 = open("STS-books-20190109.csv", "w")
fo6.write("State, School, LastName, FirstName, Age, ProjectTitle\n")
for k,v in mapProject.items():
    project_title = " ".join(v).replace(",", " ")
    state, school, student = k.split("#")
    last, first, age = student.split("_")
    fo6.write("%s, %s, %s, %s, %s, %s\n" % (state, school, last, first, age, project_title))
fo6.close()
```
